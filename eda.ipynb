```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

crop_agriculture = pd.read_csv("crop.csv")
```

```python
crop_agriculture.head()
```

```python
crop_agriculture.info()
```

```python
crop_agriculture["Area"].unique()
```

```python
plt.figure(figsize=(15,6))
top_10 = crop_agriculture.groupby("Area")["Production in Kilotonnes"].sum().nlargest(10).reset_index()
sns.barplot(x="Area",y="Production in Kilotonnes",data=top_10)
```

```python
plt.figure(figsize=(18,6))
top_10 = crop_agriculture.groupby("Area")["Production in Kilotonnes"].sum().nsmallest(10).reset_index()
sns.barplot(x="Area",y="Production in Kilotonnes",data=top_10)
```

```python
plt.figure(figsize=(18,6))
top_10 = crop_agriculture.groupby("Item")["Production in Kilotonnes"].sum().nlargest(10).reset_index()
sns.barplot(x="Item",y="Production in Kilotonnes",data=top_10)
```
